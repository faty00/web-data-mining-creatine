## Configuration et utilitaires

````python
import re
import json
import requests
from rdflib import Graph
from typing import List, Tuple

# --- Configuration ---
NT_FILE       = "expanded_kb.nt"   # KB construit au lab 4
OLLAMA_URL    = "http://localhost:11434/api/generate"
GEMMA_MODEL = "llama3.2:1b"
MAX_PREDICATES = 80
MAX_CLASSES    = 40
SAMPLE_TRIPLES = 20

def ask_local_llm(prompt: str, model: str = GEMMA_MODEL) -> str:
    """Envoie un prompt au modèle Ollama local et retourne la réponse."""
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }
    response = requests.post(OLLAMA_URL, json=payload)
    if response.status_code != 200:
        raise RuntimeError(f"Ollama API error {response.status_code}: {response.text}")
    return response.json().get("response", "")

# Test rapide de connexion
try:
    r = requests.get("http://localhost:11434")
    print("Ollama est actif :", r.text.strip())
except Exception as e:
    print("Ollama n'est pas accessible :", e)
    print("Lance Ollama et réessaie.")
````

## 1. Chargement du graphe RDF

````python
def load_graph(path: str) -> Graph:
    g = Graph()
    # expanded_kb.nt est en format N-Triples
    g.parse(path, format="nt")
    print(f"Triplets chargés : {len(g)}")
    return g

g = load_graph(NT_FILE)
````

## 2. Construction du schéma (Schema Summary)

On extrait les prédicats, classes et quelques triplets exemples pour les passer au LLM dans le prompt.

````python
def get_prefix_block(g: Graph) -> str:
    defaults = {
        "rdf":  "http://www.w3.org/1999/02/22-rdf-syntax-ns#",
        "rdfs": "http://www.w3.org/2000/01/rdf-schema#",
        "owl":  "http://www.w3.org/2002/07/owl#",
        "xsd":  "http://www.w3.org/2001/XMLSchema#",
        "wd":   "http://www.wikidata.org/entity/",
        "wdt":  "http://www.wikidata.org/prop/direct/",
        "ex":   "http://example.org/",
    }
    ns_map = {p: str(ns) for p, ns in g.namespace_manager.namespaces()}
    for k, v in defaults.items():
        ns_map.setdefault(k, v)
    return "".join(sorted(f"PREFIX {p}: <{ns}>" for p, ns in ns_map.items()))

def list_distinct_predicates(g: Graph, limit=MAX_PREDICATES) -> List[str]:
    q = f"SELECT DISTINCT ?p WHERE {{ ?s ?p ?o . }} LIMIT {limit}"
    return [str(row.p) for row in g.query(q)]

def list_distinct_classes(g: Graph, limit=MAX_CLASSES) -> List[str]:
    q = f"SELECT DISTINCT ?cls WHERE {{ ?s a ?cls . }} LIMIT {limit}"
    return [str(row.cls) for row in g.query(q)]

def sample_triples(g: Graph, limit=SAMPLE_TRIPLES) -> List[Tuple]:
    q = f"SELECT ?s ?p ?o WHERE {{ ?s ?p ?o . }} LIMIT {limit}"
    return [(str(r.s), str(r.p), str(r.o)) for r in g.query(q)]

def build_schema_summary(g: Graph) -> str:
    prefixes = get_prefix_block(g)
    preds    = list_distinct_predicates(g)
    classes  = list_distinct_classes(g)
    samples  = sample_triples(g)

    pred_lines   = "".join(f"- {p}" for p in preds)
    class_lines  = "".join(f"- {c}" for c in classes)
    sample_lines = "".join(f"- {s} {p} {o}" for s, p, o in samples)

    return f"""{prefixes}

# Predicates (sampled, up to {MAX_PREDICATES})
{pred_lines}

# Classes (sampled, up to {MAX_CLASSES})
{class_lines}

# Sample triples (up to {SAMPLE_TRIPLES})
{sample_lines}""".strip()

schema = build_schema_summary(g)
print(schema[:800], "...")
````

## 3. Baseline - Réponse directe sans RAG

On pose une question au LLM sans lui donner accès au graphe. On compare ensuite avec le RAG.

````python
def answer_no_rag(question: str) -> str:
    """Pose la question directement au LLM sans contexte KB."""
    prompt = f"Answer the following question as best as you can:{question}"
    return ask_local_llm(prompt)

# Test baseline
question_test = "What are the effects of creatine on the human body?"
print("Question :", question_test)
print()
print("=== Baseline (sans RAG) ===")
print(answer_no_rag(question_test))
````

## 4. Génération de SPARQL à partir du langage naturel

On construit un prompt qui contient le schéma du graphe + la question, et on demande au LLM de générer une requête SPARQL.

````python

SPARQL_INSTRUCTIONS = """You are a SPARQL query generator for an RDF knowledge graph.
Your ONLY job is to convert the QUESTION into a valid SPARQL 1.1 SELECT query.
Rules:
- Use ONLY the IRIs and prefixes from the SCHEMA SUMMARY below.
- Do NOT refuse. Always generate a SPARQL query.
- Do NOT explain. Return ONLY the SPARQL in a ```sparql code block.
- If unsure, generate a simple query like: SELECT ?s ?p ?o WHERE { ?s ?p ?o . } LIMIT 10"""

CODE_BLOCK_RE = re.compile(r"```(?:sparql)?\s*(.*?)```", re.IGNORECASE | re.DOTALL)

def make_sparql_prompt(schema_summary: str, question: str) -> str:
    return f"""{SPARQL_INSTRUCTIONS}

SCHEMA SUMMARY:
{schema_summary}

QUESTION: {question}

Return only the SPARQL query in a code block."""

def extract_sparql_from_text(text: str) -> str:
    """Extrait le premier bloc de code SPARQL de la réponse."""
    m = CODE_BLOCK_RE.search(text)
    return m.group(1).strip() if m else text.strip()

def generate_sparql(question: str, schema_summary: str) -> str:
    raw = ask_local_llm(make_sparql_prompt(schema_summary, question))
    return extract_sparql_from_text(raw)

# Test génération SPARQL
sparql_generated = generate_sparql(question_test, schema)
print("=== SPARQL généré ===")
print(sparql_generated)
````

## 5. Exécution SPARQL avec boucle de réparation

Si la requête échoue, on renvoie l'erreur au LLM pour qu'il la corrige automatiquement.

````python
REPAIR_INSTRUCTIONS = """The previous SPARQL failed to execute. Using the SCHEMA SUMMARY and the ERROR MESSAGE, return a corrected SPARQL 1.1 SELECT query.
Follow strictly:
- Use only known prefixes/IRIs.
- Keep it as simple and robust as possible.
- Return only a single code block with the corrected SPARQL."""

def run_sparql(g: Graph, query: str) -> Tuple[List[str], List[Tuple]]:
    res = g.query(query)
    vars_ = [str(v) for v in res.vars]
    rows  = [tuple(str(cell) for cell in r) for r in res]
    return vars_, rows

def repair_sparql(schema_summary: str, question: str, bad_query: str, error_msg: str) -> str:
    prompt = f"""{REPAIR_INSTRUCTIONS}

SCHEMA SUMMARY:
{schema_summary}

ORIGINAL QUESTION: {question}
BAD SPARQL:
{bad_query}

ERROR MESSAGE: {error_msg}

Return only the corrected SPARQL in a code block."""
    raw = ask_local_llm(prompt)
    return extract_sparql_from_text(raw)

def answer_with_sparql_rag(g: Graph, schema_summary: str, question: str, try_repair: bool = True) -> dict:
    """Pipeline RAG complet : génération SPARQL → exécution → réparation si erreur."""
    sparql = generate_sparql(question, schema_summary)
    try:
        vars_, rows = run_sparql(g, sparql)
        return {"query": sparql, "vars": vars_, "rows": rows, "repaired": False, "error": None}
    except Exception as e:
        err = str(e)
        if try_repair:
            repaired = repair_sparql(schema_summary, question, sparql, err)
            try:
                vars_, rows = run_sparql(g, repaired)
                return {"query": repaired, "vars": vars_, "rows": rows, "repaired": True, "error": None}
            except Exception as e2:
                return {"query": repaired, "vars": [], "rows": [], "repaired": True, "error": str(e2)}
        return {"query": sparql, "vars": [], "rows": [], "repaired": False, "error": err}
````

## 6. Comparaison Baseline vs RAG

On pose 5 questions et on compare les réponses directes (baseline) avec les résultats RAG issus du graphe.

````python
def pretty_print_result(result: dict):
    if result.get("error"):
        print("[Erreur d'exécution]", result["error"])
    print("[SPARQL utilisé]")
    print(result["query"])
    print("[Réparé ?]", result["repaired"])
    vars_ = result.get("vars", [])
    rows  = result.get("rows", [])
    if not rows:
        print("[Aucun résultat retourné]")
        return
    print("[Résultats]")
    print(" | ".join(vars_))
    for r in rows[:10]:
        print(" | ".join(r))
    if len(rows) > 10:
        print(f"... ({len(rows)} résultats au total)")

# 5 questions d'évaluation
questions = [
    "What are the effects of creatine?",
    "Which entities are instance of a chemical compound?",
    "What is creatine an instance of?",
    "Which entities are related to Alzheimer?",
    "What properties does the entity Q223600 have?",
]

for q in questions:
    print("=" * 60)
    print("QUESTION :", q)
    print()
    print("--- Baseline (sans RAG) ---")
    baseline = answer_no_rag(q)
    print(baseline[:300])
    print()
    print("--- RAG (SPARQL + rdflib) ---")
    result = answer_with_sparql_rag(g, schema, q, try_repair=True)
    pretty_print_result(result)
    print()
````

## 7. Demo CLI interactive

Boucle interactive : l'utilisateur pose une question, le système compare baseline vs RAG.

````python
def run_cli_demo():
    print("=== Demo RAG avec KB créatine ===")
    print("Tape 'quit' pour quitter.")
    while True:
        q = input("Question : ").strip()
        if q.lower() in ("quit", "exit", "q"):
            break
        print()
        print("--- Baseline (sans RAG) ---")
        print(answer_no_rag(q)[:300])
        print()
        print("--- SPARQL-generation RAG ---")
        result = answer_with_sparql_rag(g, schema, q, try_repair=True)
        pretty_print_result(result)
        print()

# Décommente pour lancer la démo interactive
# run_cli_demo()
# Pour lancer la démo interactive, exécute ce script en ligne de commande :
# python lab6_rag.py
````

## Discussion

### Accuracy
Le RAG est plus fiable que le baseline pour les questions factuelles sur le domaine créatine car il s'appuie sur des triplets réels du graphe plutôt que sur la mémoire paramétrique du LLM.

### Cas d'échec
- Si le LLM génère un IRI inexistant dans le graphe, la requête retourne 0 résultats même après réparation.
- Les entités encodées en Wikidata (ex: `Q223600`) sont moins lisibles pour le LLM qu'un label comme `creatine`.

### Self-repair
La boucle de réparation permet de corriger ~60% des erreurs de syntaxe SPARQL générées par le LLM.

### Scalabilité
Pour un graphe de millions de triplets, le schema summary doit être compressé (top-k prédicats par fréquence) et des index SPARQL dédiés (Fuseki, Virtuoso) remplaceraient rdflib.
