### 1.2 The Extraction Pipeline

On utilise `trafilatura` pour extraire le contenu principal de chaque page.
Une fonction vérifie que le texte est utile (> 500 mots) avant de sauvegarder.

```python
urls = [
    "https://www.protegez-vous.ca/sante-et-alimentation/supplements-proteines/le-boom-de-la-creatine-ce-que-vous-devriez-savoir-sur-ce-complement-alimentaire",
    "https://www.nutritionniste-paris.com/2025/09/25/risque-creatine/",
    "https://www.topmuscu.fr/dernieres-etudes-sur-la-creatine-monohydrate-synthese-des-recherches-2024%E2%80%912025/",
    "https://www.passeportsante.net/fr/Solutions/PlantesSupplements/Fiche.aspx?doc=creatine_ps",
    "https://www.nationalgeographic.fr/sciences/2025/02/sante-performance-anxiete-sommeil-la-creatine-un-complement-alimentaire-bon-pour-les-muscles-et-pour-le-cerveau"
]
```

```python
import urllib.robotparser

def can_fetch(url):
    rp = urllib.robotparser.RobotFileParser()
    domain = "/".join(url.split("/")[:3])
    rp.set_url(domain + "/robots.txt")
    
    try:
        rp.read()
        return rp.can_fetch("*", url)
    except:
        return False
```

```python
import httpx
import trafilatura
import json
import time

def extract_main_content(url):
    if not can_fetch(url):
        print(f"Blocked by robots.txt: {url}")
        return None
    try:
        response = httpx.get(url, timeout=20, follow_redirects=True)
        text = trafilatura.extract(response.text)
    except Exception as e:
        print(f"Erreur : {e}")
        return None

    if not text:
        print(f"{url} : Aucun texte trouvé.")
        return None

    if len(text.split()) <= 500:
        print(f"{url} : Pas assez de contenu.")
        return None

    data = {'url': url, 'word_count': len(text.split()), 'text': text}
    return json.dumps(data, ensure_ascii=False)

def create_jsonl_file(urls, filename='crawler_output.jsonl'):
    with open(filename, 'w', encoding='utf-8') as f:
        for url in urls:
            time.sleep(1)
            result = extract_main_content(url)
            if result is not None:
                word_count = json.loads(result)['word_count']
                print(f"OK ({word_count} mots) : {url[:60]}...")
                f.write(result + '\n')
            else:
                print(f"IGNORÉ : {url[:60]}...")

create_jsonl_file(urls)
```

## 5. Phase 2: Information Extraction

### 2.1 Named Entity Recognition (NER)

On charge le modèle spaCy français et on extrait les entités nommées (PER, ORG, LOC, DATE).

```python
!python -m spacy download fr_core_news_sm
```

```python
import spacy
import csv
import pandas as pd

nlp = spacy.load("fr_core_news_sm")

def read_jsonl(filename='crawler_output.jsonl'):
    """Lit le fichier JSONL ligne par ligne."""
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            yield json.loads(line)

types_utiles = {'PER', 'ORG', 'LOC', 'DATE'}
```

```python
# Extraire et afficher toutes les entités nommées
for record in read_jsonl():
    url = record['url']
    text = record['text']
    doc = nlp(text)
    print(f"Entités nommées pour l'URL: {url}")
    for ent in doc.ents:
        print(f"  {ent.text} - {ent.label_}")
    print()
```

```python
# Exporter les entités filtrées dans extracted_knowledge.csv
with open('extracted_knowledge.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['URL', 'Entité', 'Type'])
    for record in read_jsonl():
        url = record['url']
        text = record['text']
        doc = nlp(text)
        seen = set()  # éviter les doublons
        for ent in doc.ents:
            if ent.label_ in types_utiles:
                clean_text = ent.text.strip().lower()  # nettoyage
                if clean_text not in seen:
                    seen.add(clean_text)
                    writer.writerow([url, clean_text, ent.label_])

df = pd.read_csv('extracted_knowledge.csv')
print(f"Entités extraites : {len(df)}")
print()
print("Distribution par type :")
print(df['Type'].value_counts())
```

### 2.2 Introduction to Relations

#### Co-occurrence : entités dans la même phrase

On cherche les phrases contenant au moins 2 entités nommées.

```python
for record in read_jsonl():
    url = record["url"]
    text = record["text"]
    doc = nlp(text)
    for sent in doc.sents:
        ents = [ent for ent in sent.ents if ent.label_ in types_utiles]
        if len(ents) >= 2:
            print(f"URL : {url[:60]}...")
            print(f"Phrase : {sent.text.strip()}")
            for ent in ents:
                print(f"  - {ent.text} ({ent.label_})")
            print()
```

#### Triplets simples : entité + verbe + entité

On cherche un verbe connectant deux entités dans la même phrase.

```python
for record in read_jsonl():
    url = record["url"]
    text = record["text"]
    doc = nlp(text)
    for sent in doc.sents:
        ents = [ent for ent in sent.ents if ent.label_ in types_utiles]
        if len(ents) >= 2:
            for token in sent:
                if token.pos_ == "VERB":
                    subject, obj = ents[0], ents[1]
                    verb = token.text
                    print(f"({subject.text}) --[{verb}]--> ({obj.text})")
```

#### Dependency Parsing : nsubj / dobj

Méthode avancée : on utilise les tags de dépendance syntaxique pour trouver le sujet (`nsubj`) et l'objet direct (`dobj`) d'un verbe.

> **Note** : Cette méthode donne peu de résultats en français car la structure des phrases est souvent plus complexe (attributs, prépositions, passifs). Le modèle `fr_core_news_sm` détecte moins bien les dépendances que `en_core_web_trf`. C'est pourquoi la méthode simple ci-dessus est plus productive pour ce corpus.

```python
for record in read_jsonl():
    url = record["url"]
    text = record["text"]
    doc = nlp(text)
    for sent in doc.sents:
        for token in sent:
            if token.pos_ == "VERB":
                subject = None
                obj = None
                for child in token.children:
                    if child.dep_ == "nsubj" and child.ent_type_ in types_utiles:
                        subject = child
                    if child.dep_ == "dobj" and child.ent_type_ in types_utiles:
                        obj = child
                if subject is not None and obj is not None:
                    print(f"({subject.text}) --[{token.lemma_}]--> ({obj.text}) | {url[:50]}...")
```
