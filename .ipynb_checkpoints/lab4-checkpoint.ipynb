1. Prepare cleaned texts

````python
urls = [
    "https://www.protegez-vous.ca/sante-et-alimentation/supplements-proteines/le-boom-de-la-creatine-ce-que-vous-devriez-savoir-sur-ce-complement-alimentaire",
    "https://www.nutritionniste-paris.com/2025/09/25/risque-creatine/",
    "https://www.topmuscu.fr/dernieres-etudes-sur-la-creatine-monohydrate-synthese-des-recherches-2024%E2%80%912025/",
    "https://www.passeportsante.net/fr/Solutions/PlantesSupplements/Fiche.aspx?doc=creatine_ps",
    "https://www.nationalgeographic.fr/sciences/2025/02/sante-performance-anxiete-sommeil-la-creatine-un-complement-alimentaire-bon-pour-les-muscles-et-pour-le-cerveau"
]


import urllib.robotparser

def can_fetch(url):
    rp = urllib.robotparser.RobotFileParser()
    domain = "/".join(url.split("/")[:3])
    rp.set_url(domain + "/robots.txt")
    
    try:
        rp.read()
        return rp.can_fetch("*", url)
    except:
        return False
````

````python
import httpx
import trafilatura
import json
import time

def extract_main_content(url):
    if not can_fetch(url):
        print(f"Blocked by robots.txt: {url}")
        return None
    try:
        response = httpx.get(url, timeout=20, follow_redirects=True)
        text = trafilatura.extract(response.text)
    except Exception as e:
        print(f"Erreur : {e}")
        return None

    if not text:
        print(f"{url} : Aucun texte trouvé.")
        return None

    if len(text.split()) <= 500:
        print(f"{url} : Pas assez de contenu.")
        return None

    data = {'url': url, 'word_count': len(text.split()), 'text': text}
    return json.dumps(data, ensure_ascii=False)
    
# Create a JSONL file with the results
def create_jsonl_file(urls, filename='crawler_output.jsonl'):
    with open(filename, 'w', encoding='utf-8') as f:
        for url in urls:
            time.sleep(1)
            result = extract_main_content(url)
            if result is not None:
                word_count = json.loads(result)['word_count']
                print(f"OK ({word_count} mots) : {url[:60]}...")
                f.write(result + '\n')
            else:
                print(f"IGNORÉ : {url[:60]}...")

create_jsonl_file(urls)
````

2. Named Entity Recognition (NER) with spaCy

````python
import spacy
import csv
import pandas as pd

nlp = spacy.load("fr_core_news_sm")

def read_jsonl(filename='crawler_output.jsonl'):
    """Lit le fichier JSONL ligne par ligne."""
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            yield json.loads(line)
````

````python
# Extraire et afficher toutes les entités nommées
for record in read_jsonl():
    url = record['url']
    text = record['text']
    doc = nlp(text)
    print(f"Entités nommées pour l'URL: {url}")
    for ent in doc.ents:
        print(f"  {ent.text} - {ent.label_}") # Display the entity text and its label
    print("\n")

# Filter useful entities PERSON, ORG (Organisation), GPE (Location), and DATE
types_utiles = {'PER', 'ORG', 'LOC', 'DATE'}

def is_valid_entity(ent):
    text = ent.text.strip()
    if len(text) <= 3:
        return False
    if "'" in text or "\u2019" in text:
        return False
    if not text[0].isupper():
        return False
    # Virer les entités dont le token principal est un verbe
    if ent.root.pos_ in {"VERB", "ADV", "ADJ", "DET"}:
        return False
    return True
````

````python
# Exporter les entités filtrées dans extracted_knowledge.csv
with open('extracted_knowledge.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['URL', 'Entité', 'Type'])
    for record in read_jsonl():
        url = record['url']
        text = record['text']
        doc = nlp(text)
        seen = set()  # éviter les doublons
        for ent in doc.ents:
            if ent.label_ in types_utiles:
                clean_text = ent.text.strip().lower()  # nettoyage
                if clean_text not in seen:
                    seen.add(clean_text)
                    writer.writerow([url, clean_text, ent.label_])

df = pd.read_csv('extracted_knowledge.csv')
print(f"Entités extraites : {len(df)}")
print()
print("Distribution par type :")
print(df['Type'].value_counts())
````

````python
# Go through each sentence and see if there are multiple named entities
for record in read_jsonl():
    url = record["url"]
    doc = nlp(record["text"])
    for sent in doc.sents:
        ents = [ent for ent in sent.ents 
                if ent.label_ in types_utiles and is_valid_entity(ent)]
        if len(ents) >= 2:
            print(f"URL : {url[:60]}...")
            print(f"Phrase : {sent.text}")
            for ent in ents:
                print(f"  - {ent.text} ({ent.label_})")
            print()
````

````python
import urllib.parse
from typing import Set, Tuple

def text_to_uri(text: str, base: str = 'http://example.org/resource/') -> str:
    key = urllib.parse.quote(text.strip().replace(' ', '_'), safe='')
    return base + key

def pred_to_uri(lemma: str, base: str = 'http://example.org/property/') -> str:
    # camelCase : première lettre minuscule
    key = urllib.parse.quote(lemma.strip().replace(' ', '_'), safe='')
    return base + key[0].lower() + key[1:]

def escape_literal(s: str) -> str:
    return s.replace('\\', '\\\\').replace('"', '\\"')

triplet_set: Set[Tuple[str, str, str]] = set()
labels_set:  Set[Tuple[str, str]] = set()

for record in read_jsonl():
    doc = nlp(record['text'])
    for sent in doc.sents:
        # Entités valides dans cette phrase
        ents = [e for e in sent.ents if e.label_ in types_utiles and is_valid_entity(e)]
        if len(ents) < 2:
            continue
                    # We are looking for a verb in the sentence.

        for token in sent:
            if token.pos_ == 'VERB':
                subj = ents[0]
                obj  = ents[1]
                s_uri = text_to_uri(subj.text)
                o_uri = text_to_uri(obj.text)
                p_uri = pred_to_uri(token.lemma_)
                triplet_set.add((s_uri, p_uri, o_uri))
                labels_set.add((s_uri, subj.text.strip()))
                labels_set.add((o_uri, obj.text.strip()))

print(f"Triplets extraits  : {len(triplet_set)}")
print(f"Entités avec label : {len(labels_set)}")
print()
print("Exemples de triplets :")
for s, p, o in list(triplet_set)[:8]:
    s_l = s.split('/')[-1]
    p_l = p.split('/')[-1]
    o_l = o.split('/')[-1]
    print(f"  ({s_l}) --[{p_l}]--> ({o_l})")
````

````python
# Triplets manuels sur le domaine créatine
# Justification : le NER fr_core_news_sm ne reconnaît pas les noms communs
# scientifiques comme entités nommées (limitation connue du petit modèle)

manual_triplets = [
    # Créatine - propriétés de base
    ("creatine",            "isA",              "dietary_supplement"),
    ("creatine",            "isA",              "chemical_compound"),
    ("creatine",            "hasEffect",        "muscle"),
    ("creatine",            "hasEffect",        "brain"),
    ("creatine",            "hasEffect",        "sleep"),
    ("creatine",            "hasEffect",        "depression"),
    ("creatine",            "hasEffect",        "anxiety"),
    ("creatine",            "storedIn",         "muscle"),
    ("creatine",            "storedIn",         "brain"),
    ("creatine",            "metabolizedTo",    "phosphocreatine"),
    ("creatine",            "convertedTo",      "creatinine"),
    ("creatine",            "synthesizedFrom",  "amino_acid"),
    # Chaîne énergétique
    ("phosphocreatine",     "regenerates",      "ATP"),
    ("ATP",                 "isA",              "energy_molecule"),
    ("ATP",                 "usedBy",           "muscle"),
    ("ATP",                 "usedBy",           "brain"),
    ("phosphocreatine",     "storedIn",         "muscle"),
    # Chercheurs
    ("Scott_Forbes",        "affiliatedWith",   "Université_de_Brandon"),
    ("Scott_Forbes",        "locatedIn",        "Manitoba"),
    ("Scott_Forbes",        "researchedOn",     "creatine"),
    ("Scott_Forbes",        "publishedIn",      "Scientific_Reports"),
    ("Abbie_E._Smith-Ryan", "affiliatedWith",   "Chapel_Hill"),
    ("Abbie_E._Smith-Ryan", "researchedOn",     "creatine"),
    ("Abbie_E._Smith-Ryan", "publishedIn",      "Nutrients"),
    # Institutions et lieux
    ("Université_de_Brandon","locatedIn",       "Manitoba"),
    ("Manitoba",            "locatedIn",        "Canada"),
    ("Chapel_Hill",         "locatedIn",        "Canada"),
    # Effets sur la santé
    ("creatine",            "improves",         "muscle_strength"),
    ("creatine",            "improves",         "cognitive_function"),
    ("creatine",            "reduces",          "depression"),
    ("creatine",            "reduces",          "anxiety"),
    ("creatine",            "improves",         "sleep_quality"),
    ("creatine",            "safeFor",          "kidney"),
    ("creatine",            "studiedBy",        "International_Society_of_Sports_Nutrition"),
    ("creatine",            "hasStatus",        "Generally_Recognized_as_Safe"),
    # Maladies et conditions
    ("Alzheimer",           "isA",              "neurodegenerative_disease"),
    ("creatine",            "investigatedFor",  "Alzheimer"),
    ("depression",          "isA",              "mental_disorder"),
    ("anxiety",             "isA",              "mental_disorder"),
    ("kidney",              "metabolizes",      "creatinine"),
    # Journaux scientifiques
    ("Scientific_Reports",  "isA",              "scientific_journal"),
    ("Nutrients",           "isA",              "scientific_journal"),
    ("MDPI",                "publishes",        "Nutrients"),
    # Cellules immunitaires
    ("Natural_Killer",      "isA",              "immune_cell"),
    ("creatine",            "hasEffect",        "Natural_Killer"),
]

# Ajout dans triplet_set
for s, p, o in manual_triplets:
    s_uri = text_to_uri(s)
    p_uri = pred_to_uri(p)
    o_uri = text_to_uri(o)
    triplet_set.add((s_uri, p_uri, o_uri))
    labels_set.add((s_uri, s.replace("_", " ")))
    labels_set.add((o_uri, o.replace("_", " ")))

print(f"Triplets après ajout manuel : {len(triplet_set)}")
print(f"Entités après ajout manuel  : {len(labels_set)}")
````

````python
def save_nt(filename='graph_knowledge.nt'):
    with open(filename, 'w', encoding='utf-8') as f:
        # Triplets (URI URI URI)
        for s, p, o in sorted(triplet_set):
            f.write(f'<{s}> <{p}> <{o}> .\n')
        # Labels (URI rdfs:label "texte"@fr)
        rdfs_label = 'http://www.w3.org/2000/01/rdf-schema#label'
        for uri, label in sorted(labels_set):
            esc = escape_literal(label)
            f.write(f'<{uri}> <{rdfs_label}> "{esc}"@fr .\n')

save_nt()

# Stats finales Step 1
entities = {s for s,p,o in triplet_set} | {o for s,p,o in triplet_set}
predicates = {p for s,p,o in triplet_set}

print("=== Statistiques KB initiale ===")
print(f"  Triplets   : {len(triplet_set)}")
print(f"  Entités    : {len(entities)}")
print(f"  Prédicats  : {len(predicates)}")
print()
req_triplets = len(triplet_set) >= 100
req_entities = len(entities) >= 50
print(f"  >= 100 triplets : {'OK' if req_triplets else 'PAS ENCORE'}")
print(f"  >= 50 entités   : {'OK' if req_entities else 'PAS ENCORE'}")
print()
print("Fichier sauvegardé : graph_knowledge.nt")
````

## Step 2 --- Entity Linking with Open Knowledge Bases

### 1. Connection to the Wikidata API

````python
import requests
import pandas as pd
from rdflib import Graph, Namespace
from rdflib.namespace import OWL, RDF, RDFS

api_url = "https://www.wikidata.org/w/api.php"
EX = Namespace("http://example.org/")
WD = Namespace("http://www.wikidata.org/entity/")
````

### 2. Fonction de recherche Wikidata

````python
def get_best_match(entity_name):
    params = {
        "action": "wbsearchentities",
        "format": "json",
        "language": "en",
        "search": entity_name
    }
    headers = {
        "User-Agent": "TP-KB-Construction/1.0"
    }
    response = requests.get(api_url, params=params, headers=headers)
    data = response.json()

    if "search" in data and len(data["search"]) > 0:
        best = data["search"][0]
        return {
            "wikidata_id": best["id"],
            "label": best["label"],
            "description": best.get("description", ""),
            "confidence": best["match"].get("score", 1.0)
        }
    else:
        return None
````

### 3. Liste des entités privées

Entités extraites par le NER (Step 1) + entités du domaine créatine non détectées par spaCy (limitation du petit modèle fr_core_news_sm).

````python
import re

# Lire les entités directement depuis graph_knowledge.nt
def load_entities_from_nt(filename='graph_knowledge.nt'):
    entities = set()
    uri_pattern = re.compile(r'<http://example\.org/resource/([^>]+)>')
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            matches = uri_pattern.findall(line)
            for match in matches:
                # Décoder l'URI en texte lisible
                from urllib.parse import unquote
                label = unquote(match).replace('_', ' ')
                entities.add(label)
    return sorted(entities)

ner_entities = load_entities_from_nt()

# Entités du domaine créatine non détectées par spaCy
domain_entities = [
    "creatine",
    "phosphocreatine",
    "ATP",
    "muscle",
    "brain",
    "kidney",
    "depression",
    "anxiety",
    "sleep",
]

private_entities = ner_entities + domain_entities
print(f"Total entités à aligner : {len(private_entities)}")
print("Entités chargées depuis graph_knowledge.nt :")
for e in ner_entities:
    print(f"  {e}")
````

### 4. Entity linking + mapping table

````python
alignment_graph = Graph()
alignment_graph.bind("ex", EX)
alignment_graph.bind("owl", OWL)
alignment_graph.bind("wd", WD)
alignment_graph.bind("rdfs", RDFS)

mapping_data = []

for entity in private_entities:
    match = get_best_match(entity)

    # URI locale : espaces remplacés par underscore
    entity_uri = EX[entity.replace(" ", "_")]

    if match:
        wikidata_uri = WD[match["wikidata_id"]]
        confidence = match["confidence"]

        # Alignement owl:sameAs
        alignment_graph.add((entity_uri, OWL.sameAs, wikidata_uri))

        print(f"  FOUND  {entity:45s} -> {match['wikidata_id']} ({match['label']})")
    else:
        wikidata_uri = None
        confidence = 0.0

        # Nouvelle entité : définition ontologique
        alignment_graph.add((entity_uri, RDF.type, EX.CreatineRelatedConcept))

        print(f"  NEW    {entity:45s} -> définie localement")

    mapping_data.append({
        "Private Entity": entity,
        "External URI": f"http://www.wikidata.org/entity/{match['wikidata_id']}" if match else "NEW ENTITY",
        "Confidence": round(confidence, 2)
    })

print(f"\nTotal : {len(mapping_data)} entités traitées")
````

### 5. Définition ontologique

````python
# Hiérarchie de classes
alignment_graph.add((EX.CreatineSupplement, RDFS.subClassOf, EX.DietarySupplement))
alignment_graph.add((EX.DietarySupplement,  RDFS.subClassOf, EX.Substance))
alignment_graph.add((EX.Researcher,         RDFS.subClassOf, EX.Person))
alignment_graph.add((EX.University,         RDFS.subClassOf, EX.Organization))

# Propriétés avec domain et range
alignment_graph.add((EX.affiliatedWith, RDFS.domain, EX.Person))
alignment_graph.add((EX.affiliatedWith, RDFS.range,  EX.Organization))
alignment_graph.add((EX.hasEffect,      RDFS.domain, EX.CreatineSupplement))
alignment_graph.add((EX.hasEffect,      RDFS.range,  EX.BiologicalProcess))
alignment_graph.add((EX.studiedBy,      RDFS.domain, EX.CreatineSupplement))
alignment_graph.add((EX.studiedBy,      RDFS.range,  EX.Researcher))
````

### 6. Export

````python
# Export fichier d'alignement
alignment_graph.serialize("alignment.ttl", format="turtle")
print("alignment.ttl sauvegardé")

# Export mapping table
df = pd.DataFrame(mapping_data)
df.to_csv("mapping_table.csv", index=False)
print("mapping_table.csv sauvegardé")
print()
print(df.to_string(index=False))
````

## Step 3 — Predicate Alignment Using SPARQL

### 1. Setup SPARQLWrapper

````python
!pip install sparqlwrapper
````

````python
from SPARQLWrapper import SPARQLWrapper, JSON
import time

sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
sparql.setReturnFormat(JSON)
sparql.addCustomHttpHeader("User-Agent", "TP-KB-Construction/1.0")
````

### 2. Méthode 1 — Propriétés existantes entre deux entités alignées

Pour chaque triplet `(s, p, o)` dont `s` et `o` sont alignés dans Wikidata, on cherche quelles propriétés existent déjà entre `wiki(s)` et `wiki(o)`.

````python
def get_properties_between(qs, qo):
    """Liste toutes les propriétés Wikidata entre deux entités alignées."""
    query = f"""
    SELECT ?property ?propertyLabel WHERE {{
      ?property a wikibase:Property .
      ?property rdfs:label ?propertyLabel .
      wd:{qs} ?property wd:{qo} .
      FILTER(LANG(?propertyLabel) = "en")
    }}
    """
    sparql.setQuery(query)
    try:
        results = sparql.query().convert()
        return [
            (r["property"]["value"].split("/")[-1], r["propertyLabel"]["value"])
            for r in results["results"]["bindings"]
        ]
    except Exception as e:
        print(f"  Erreur : {e}")
        return []

# Triplets dont les deux entités sont alignées dans Wikidata
triplets_alignes = [
    ("locatedIn",      "Manitoba",       "Canada",               "Q1948",      "Q16"),
    ("affiliatedWith", "Scott_Forbes",   "Université_de_Brandon","Q112500633",  "Q4957110"),
    ("metabolizedTo",  "creatine",       "phosphocreatine",      "Q223600",    "Q1984607"),
    ("publishedIn",    "Scott_Forbes",   "Scientific_Reports",   "Q112500633", "Q2261792"),
]

print("=" * 65)
print("MÉTHODE 1 — Propriétés existantes entre entités alignées")
print("=" * 65)

for pred, s, o, qs, qo in triplets_alignes:
    print(f"\nex:{pred}  ({s} -> {o})")
    print(f"  Requête : wd:{qs} ?p wd:{qo}")
    props = get_properties_between(qs, qo)
    if props:
        for pid, plabel in props:
            print(f"  -> {pid}: {plabel}")
    else:
        print("  -> Aucune propriété directe trouvée")
    time.sleep(1)
````

### 3. Méthode 2 — Recherche par mot-clé dans les labels

Pour les prédicats sans entités alignées des deux côtés, on cherche par mot-clé.

````python
def find_properties_by_keyword(keyword, limit=10):
    """Cherche des propriétés Wikidata dont le label contient un mot-clé."""
    query = f"""
    SELECT ?property ?propertyLabel WHERE {{
      ?property a wikibase:Property .
      ?property rdfs:label ?propertyLabel .
      FILTER(CONTAINS(LCASE(?propertyLabel), "{keyword.lower()}"))
      FILTER(LANG(?propertyLabel) = "en")
    }}
    LIMIT {limit}
    """
    sparql.setQuery(query)
    try:
        results = sparql.query().convert()
        return [
            (r["property"]["value"].split("/")[-1], r["propertyLabel"]["value"])
            for r in results["results"]["bindings"]
        ]
    except Exception as e:
        print(f"  Erreur : {e}")
        return []

predicates_to_search = {
    "hasEffect":   "effect",
    "studiedBy":   "researcher",
    "improves":    "improve",
    "isA":         "instance",
    "storedIn":    "located",
}

print("=" * 65)
print("MÉTHODE 2 — Recherche par mot-clé")
print("=" * 65)

for pred, keyword in predicates_to_search.items():
    print(f"\nex:{pred}  (recherche: '{keyword}')")
    candidates = find_properties_by_keyword(keyword, limit=5)
    for pid, plabel in candidates:
        print(f"  -> {pid}: {plabel}")
    time.sleep(1)
````

### 4. Décisions d'alignement (validation manuelle)

Après comparaison des résultats SPARQL avec nos prédicats, on choisit la propriété la plus pertinente.

- `owl:equivalentProperty` → même signification exacte
- `rdfs:subPropertyOf` → notre prédicat est plus spécifique que celui de Wikidata

````python
from rdflib import Graph, Namespace
from rdflib.namespace import OWL, RDFS

EX  = Namespace("http://example.org/")
WDT = Namespace("http://www.wikidata.org/prop/direct/")

predicate_graph = Graph()
predicate_graph.bind("ex",   EX)
predicate_graph.bind("wdt",  WDT)
predicate_graph.bind("owl",  OWL)
predicate_graph.bind("rdfs", RDFS)

# (prédicat privé, propriété Wikidata, type de relation, justification)
alignments = [
    # Méthode 1 — trouvé directement entre entités alignées
    (EX.locatedIn,      WDT.P131,  OWL.equivalentProperty, "P131: located in administrative entity"),
    (EX.affiliatedWith, WDT.P1416, OWL.equivalentProperty, "P1416: affiliation"),
    (EX.publishedIn,    WDT.P1433, OWL.equivalentProperty, "P1433: published in"),
    #(EX.metabolizedTo,  WDT.P1060, RDFS.subPropertyOf,     "P1060: pathogen transmission — plus général"),
        #(EX.metabolizedTo,  WDT.P1060, RDFS.subPropertyOf,     "P1060: pathogen transmission — plus général"),

    # Méthode 2 — trouvé par mot-clé
    (EX.hasEffect,      WDT.P1542, OWL.equivalentProperty, "P1542: has effect"),
    (EX.studiedBy,      WDT.P1450, OWL.equivalentProperty, "P1450: studied by"),
    (EX.isA,            WDT.P31,   OWL.equivalentProperty, "P31: instance of"),
    (EX.storedIn,       WDT.P276,  RDFS.subPropertyOf,     "P276: location — plus général"),
]

print(f"{'Prédicat privé':20s} {'Relation':25s} {'Wikidata':10s} {'Justification'}")
print("-" * 85)

for priv, wiki, rel, note in alignments:
    predicate_graph.add((priv, rel, wiki))
    rel_label = "owl:equivalentProperty" if rel == OWL.equivalentProperty else "rdfs:subPropertyOf"
    pid   = str(wiki).split("/")[-1]
    pname = str(priv).split("/")[-1]
    print(f"  ex:{pname:18s} {rel_label:25s} wdt:{pid:8s} {note}")

predicate_graph.serialize("predicate_alignment.ttl", format="turtle")
print(f"\npredicate_alignment.ttl sauvegardé ({len(predicate_graph)} triplets)")
````

## Step 4 — KB Expansion via SPARQL

### 1. Setup

````python
from SPARQLWrapper import SPARQLWrapper, JSON
import time

sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
sparql.setReturnFormat(JSON)
sparql.addCustomHttpHeader("User-Agent", "TP-KB-Construction/1.0")

expanded_triples = set()

# Entités core alignées (issues du Step 2)
core_entities = {
    "creatine":        "Q223600",
    "phosphocreatine": "Q1984607",
    "ATP":             "Q80863",
    "muscle":          "Q7365",
    "brain":           "Q897386",
    "kidney":          "Q9377",
    "depression":      "Q42844",
    "anxiety":         "Q154430",
    "sleep":           "Q35831",
    "Alzheimer":       "Q11081",
    "Natural_Killer":  "Q332181",
}

print(f"Entités core : {len(core_entities)}")
````

### 2. Expansion 1-hop

Pour chaque entité alignée, on récupère tous ses voisins directs.

Requête exacte du TP :
```sparql
SELECT ?p ?o WHERE { wd:Q7186 ?p ?o . } LIMIT 1000
```

````python
def sparql_1hop(qid, limit=1000):
    query = f"""
    SELECT ?p ?o WHERE {{
      wd:{qid} ?p ?o .
    }}
    LIMIT {limit}
    """
    sparql.setQuery(query)
    try:
        results = sparql.query().convert()
        return [
            (f"http://www.wikidata.org/entity/{qid}",
             r["p"]["value"],
             r["o"]["value"])
            for r in results["results"]["bindings"]
        ]
    except Exception as e:
        print(f"  Erreur {qid}: {e}")
        return []

print("Expansion 1-hop...")
print("=" * 50)

for name, qid in core_entities.items():
    triples = sparql_1hop(qid, limit=1000)
    for t in triples:
        expanded_triples.add(t)
    print(f"  {name:20s} (wd:{qid}) : +{len(triples):5d}  (total: {len(expanded_triples)})")
    time.sleep(1.5)

print(f"\nAprès 1-hop : {len(expanded_triples)} triplets")
````

### 3. Predicate-Controlled Expansion

Pour chaque propriété alignée, on récupère tous les triplets qui l'utilisent.

Requête exacte du TP :
```sparql
SELECT ?s ?p ?o WHERE { ?s wdt:P166 ?o . } LIMIT 20000
```

````python
def sparql_predicate_expand(sparql_query, label=""):
    """Expansion contrôlée : exécute une requête SPARQL et filtre les triplets wdt:"""
    sparql.setQuery(sparql_query)
    try:
        results = sparql.query().convert()
        triplets = set()
        for r in results["results"]["bindings"]:
            s = r["s"]["value"]
            p = r["p"]["value"]
            o = r["o"]["value"]
            # Garder seulement les vraies propriétés wdt: entre entités Q...
            if "prop/direct" in p and o.startswith("http://www.wikidata.org/entity/Q"):
                triplets.add((s, p, o))
        print(f"  {label} : {len(triplets)} triplets récupérés")
        return triplets
    except Exception as e:
        print(f"  Erreur {label}: {e}")
        return set()

q1 = """
SELECT ?s ?p ?o WHERE {
  ?s wdt:P31 wd:Q11173 .
  ?s ?p ?o .
  FILTER(STRSTARTS(STR(?p), "http://www.wikidata.org/prop/direct/"))
  FILTER(STRSTARTS(STR(?o), "http://www.wikidata.org/entity/Q"))
} LIMIT 20000
"""

q2 = """
SELECT ?s ?p ?o WHERE {
  ?s wdt:P31 wd:Q188055 .
  ?s ?p ?o .
  FILTER(STRSTARTS(STR(?p), "http://www.wikidata.org/prop/direct/"))
  FILTER(STRSTARTS(STR(?o), "http://www.wikidata.org/entity/Q"))
} LIMIT 20000
"""

q3 = """
SELECT ?s ?p ?o WHERE {
  ?s wdt:P31 wd:Q8054 .
  ?s ?p ?o .
  FILTER(STRSTARTS(STR(?p), "http://www.wikidata.org/prop/direct/"))
  FILTER(STRSTARTS(STR(?o), "http://www.wikidata.org/entity/Q"))
} LIMIT 20000
"""

q4 = """
SELECT ?s ?p ?o WHERE {
  ?s wdt:P279 wd:Q11173 .
  ?s ?p ?o .
  FILTER(STRSTARTS(STR(?p), "http://www.wikidata.org/prop/direct/"))
  FILTER(STRSTARTS(STR(?o), "http://www.wikidata.org/entity/Q"))
} LIMIT 20000
"""

print("Predicate-controlled expansion...")
t1 = sparql_predicate_expand(q1, "composés chimiques (P31)")
time.sleep(2)
t2 = sparql_predicate_expand(q2, "suppléments alimentaires (P31)")
time.sleep(2)
t3 = sparql_predicate_expand(q3, "protéines (P31)")
time.sleep(2)
t4 = sparql_predicate_expand(q4, "sous-classes composés chimiques (P279)")
time.sleep(2)

expanded_triples.update(t1)
expanded_triples.update(t2)
expanded_triples.update(t3)
expanded_triples.update(t4)

print(f"\nAprès predicate-controlled : {len(expanded_triples)} triplets")
````

### 4. Expansion 2-hop

Requête exacte du TP :
```sparql
SELECT ?award ?p ?o WHERE {
  ?person wdt:P166 ?award .
  ?award ?p ?o .
} LIMIT 10000
```

On adapte avec nos propriétés alignées : on part des objets trouvés via `wdt:P1542` (has effect) et on récupère leurs voisins.

````python
def sparql_2hop(prop_id, limit=10000):
    """2-hop : on part des objets liés via prop_id depuis nos entités core,
    puis on récupère leurs voisins."""
    core_values = " ".join([f"wd:{qid}" for qid in core_entities.values()])
    query = f"""
    SELECT ?obj ?p ?o WHERE {{
      VALUES ?entity {{ {core_values} }}
      ?entity wdt:{prop_id} ?obj .
      ?obj ?p ?o .
    }}
    LIMIT {limit}
    """
    sparql.setQuery(query)
    try:
        results = sparql.query().convert()
        return [
            (r["obj"]["value"],
             r["p"]["value"],
             r["o"]["value"])
            for r in results["results"]["bindings"]
        ]
    except Exception as e:
        print(f"  Erreur 2-hop P{prop_id}: {e}")
        return []

print("Expansion 2-hop...")
print("=" * 50)

# 2-hop depuis has effect (P1542) et instance of (P31)
for prop_id, label in [("P1542", "has effect"), ("P31", "instance of"), ("P1416", "affiliation")]:
    triples = sparql_2hop(prop_id, limit=10000)
    for t in triples:
        expanded_triples.add(t)
    print(f"  2-hop via P{prop_id} ({label:20s}) : +{len(triples):6d}  (total: {len(expanded_triples)})")
    time.sleep(2)

print(f"\nAprès 2-hop : {len(expanded_triples)} triplets")
````

### 5. Nettoyage

Comme demandé dans le TP :
- Supprimer les doublons
- Supprimer les URIs incohérentes
- Supprimer les prédicats internes Wikibase

````python
UNWANTED_PREFIXES = [
    "http://schema.org/",
    "http://www.w3.org/ns/prov#",
    "http://wikiba.se/ontology#",
    "http://www.mediawiki.org/",
    "http://www.wikidata.org/prop/reference",
    "http://www.wikidata.org/prop/qualifier",
]

def is_wanted(triple):
    s, p, o = triple
    for prefix in UNWANTED_PREFIXES:
        if p.startswith(prefix):
            return False
    return True

before = len(expanded_triples)
cleaned_triples = set(filter(is_wanted, expanded_triples))

print(f"Avant nettoyage : {before:,}")
print(f"Supprimés       : {before - len(cleaned_triples):,}")
print(f"Après nettoyage : {len(cleaned_triples):,}")
````

````python
from collections import Counter

# Garder seulement les 200 relations les plus fréquentes
relation_counts = Counter(p for s, p, o in expanded_triples)
print(f"Relations avant filtre : {len(relation_counts)}")

top_200_relations = {rel for rel, count in relation_counts.most_common(150)}
cleaned_triples = {(s, p, o) for s, p, o in expanded_triples if p in top_200_relations}

print(f"Relations après filtre : 200")
print(f"Triplets après filtre  : {len(cleaned_triples)}")

# Remplacer expanded_triples par la version filtrée
expanded_triples = cleaned_triples
````

### 6. Export — expanded_kb.nt

````python
def fmt(v):
    if v.startswith("http") or v.startswith("urn"):
        return f"<{v}>"
    esc = v.replace("\\", "\\\\").replace('"', '\\"')
    return f'"{esc}"'

# Charger le KB privé
private_triples = []
try:
    with open("graph_knowledge.nt", "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line and not line.startswith("#"):
                private_triples.append(line)
    print(f"KB privé : {len(private_triples)} triplets")
except FileNotFoundError:
    print("graph_knowledge.nt non trouvé")

# Export
with open("expanded_kb.nt", "w", encoding="utf-8") as f:
    for line in private_triples:
        f.write(line + "\n")
    for s, p, o in sorted(cleaned_triples):
        f.write(f"{fmt(s)} {fmt(p)} {fmt(o)} .\n")

print(f"expanded_kb.nt sauvegardé")
print(f"Total : {len(private_triples) + len(cleaned_triples):,} triplets")
````

### 7. Statistiques finales

````python
import re

with open("expanded_kb.nt", "r", encoding="utf-8") as f:
    lines = [l.strip() for l in f if l.strip() and not l.startswith("#")]

uri_pattern = re.compile(r"<([^>]+)>")
entities, predicates = set(), set()

for line in lines:
    uris = uri_pattern.findall(line)
    if len(uris) >= 3:
        entities.add(uris[0])
        predicates.add(uris[1])
        if uris[2].startswith("http"):
            entities.add(uris[2])

print("=" * 45)
print("STATISTIQUES FINALES KB")
print("=" * 45)
print(f"  Triplets  : {len(lines):>10,}")
print(f"  Entités   : {len(entities):>10,}")
print(f"  Relations : {len(predicates):>10,}")
print()
print("Contraintes du TP :")
print(f"  50k-200k triplets : {'OK' if 50000 <= len(lines) <= 200000 else 'PAS ENCORE ' + str(len(lines))}")
print(f"  5k-30k entités    : {'OK' if 5000 <= len(entities) <= 30000 else 'PAS ENCORE ' + str(len(entities))}")
print(f"  50-200 relations  : {'OK' if 50 <= len(predicates) <= 200 else str(len(predicates))}")
````
